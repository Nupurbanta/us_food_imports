```python
import pandas as pd   
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
foodimports=pd.read_excel('../data/FoodImports.xlsx', sheet_name='FoodValue', header=None)
foodimports
```

```python
#changed rows to columns
df_1 =foodimports.transpose()
df_1
```

```python
#picked only few columns
df_2 = df_1[[3,4, 21, 22, 23]]
df_2 
```

```python
#drop few rows
df_3 = df_2.drop(labels=[0,1,2,3], axis=0)
df_3
```

```python
#remane the columns
df_4 = df_3.rename(columns={3: 'Years', 4:'Total_Foods', 21: 'Animal_Food', 22: 'Plant_Foods', 23:'Beverage'})
df_4
```

```python
df_4.dtypes
```

```python
# Assuming df is your DataFrame and 'Years' is the column containing values like "1999.0"

df_4['Years'] = df_4['Years'].astype(str).str.replace('.0', '').astype(int)
df_4['Years'] = df_4['Years'].astype(object)
```

```python
df_4.dtypes
```

```python
df_4.sort_values('Years').plot(x='Years',y='Total_Foods');
df_4.sort_values('Years').plot(x='Years',y='Animal_Food');
df_4.sort_values('Years').plot(x='Years',y='Plant_Foods');
df_4.sort_values('Years').plot(x='Years',y='Beverage');
```

```python
df_4.to_csv('total_usfoodimport.csv')
```
