```python
import pandas as pd   
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
foodimports=pd.read_excel('../data/FoodImports.xlsx', sheet_name='FoodVolume', header=None)
foodimports
```

```python
#changed rows to columns
df_1 =foodimports.transpose()
df_1
```

```python
df_1.drop(columns=df_1.columns[[0, 1, 2]], inplace=True)
df_1
```

```python
 df_2 = df_1.drop(labels=[0,1,2], axis=0)
df_2
```

```python
columns_to_drop = df_2.columns[15:24]
```

```python
df_3 = df_2.drop(columns=columns_to_drop, axis=1)
```

```python
df_3
```

```python
#remane the columns
df_4 = df_3.rename(columns={3: 'Years', 4:'Live meat animals', 5: 'Meats', 6: 'Fish and shellfish ', 7:'Dairy', 8: 'Vegetables',
                            9: 'Fruits',10: 'Nuts',11: 'Coffee, tea, and spices',12: 'Grains',13: 'Vegetable oils',14: 'Sugar and candy', 
                           15: 'Cocoa and chocolate',16: 'Other edible products',17: 'Beverages'})
df_4
```

```python
# Assuming df is your DataFrame and 'Years' is the column containing values like "1999.0"

df_4['Years'] = df_4['Years'].astype(str).str.replace('.0', '').astype(int)
df_4['Years'] = df_4['Years'].astype(object)
```
