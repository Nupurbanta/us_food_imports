```python
import pandas as pd   
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
foodimports=pd.read_excel('../data/FoodImports.xlsx', sheet_name='FoodValue', header=None)
foodimports
```

```python
#changed rows to columns
df_1 =foodimports.transpose()
df_1
```

```python
#picked only few columns
df_2 = df_1[[3,4, 21, 22, 23]]
df_2
```

```python
#drop few rows
df_3 = df_2.drop(labels=[0,1,2,3], axis=0)
df_3
```

```python
#remane the columns
df_4 = df_3.rename(columns={3: 'Years', 4:'Total_Foods', 21: 'Animal_Food', 22: 'Plant_Foods', 23:'Beverage'})
df_4
```

```python
df_4['Years'] = df_4['Years'].astype(str).str.replace('.0', '').astype(int)
df_4['Years'] = df_4['Years'].astype(object)
```

```python
df_4
```

```python
df_4 = df_4.drop(columns =['Animal_Food'])
```

```python
df_4 = df_4.drop(columns =['Plant_Foods'])
```

```python
df_4 = df_4.drop(columns =['Beverage'])
```

```python
df_population=pd.read_excel('../data/Population.xlsx', sheet_name='US Population', header=None)
df_population = df_population.rename(columns={0: 'Year', 1:'Population'})
df_population["Years"] = ''
df_population["Years"].apply(str)
df_population["Years"]  = df_population["Year"].dt.year
df_population["Years"] = df_population["Years"].astype(str)
df_population["Years"] = pd.to_numeric(df_population["Years"])
df_population["Population"] = df_population["Population"].astype(float)
df_population["Population"] = pd.to_numeric(df_population["Population"])
df_6 = df_population

df_7 = df_6.drop(labels=[0], axis=0)
df_8 = df_7.drop(columns =['Year'])
```
